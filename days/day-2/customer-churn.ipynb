```python
# Drop ID column
df = df.drop("customerID", axis=1)

# Encode categorical features
df_new = pd.get_dummies(df, drop_first=True)
df_new.head()
```

```python
from sklearn.feature_selection import mutual_info_classif
import pandas as pd

X = df_new.drop( "Churn_Yes", axis=1)
y = df_new["Churn_Yes"]

mi = mutual_info_classif(X, y)

mi_series = pd.Series(mi, index=X.columns)

```

```python
mi_series = pd.Series(mi, index=X.columns)
```

```python
top_features = mi_series.sort_values(ascending=False).head(10).index
```

```python
top_features
```

```python
# removing repeated features
top_features = top_features.drop(["DeviceProtection_No internet service", "StreamingMovies_No internet service", "OnlineSecurity_No internet service"])
```

```python
top_features
```

```python
from sklearn.model_selection import train_test_split
X_sel = X[top_features]
X_train, X_test, y_train, y_test = train_test_split(X_sel, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
model = RandomForestClassifier(class_weight="balanced")
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight="balanced", max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rf = RandomForestClassifier(
    class_weight="balanced",
    n_estimators=300,
    max_depth=6,
    random_state=42
)

rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

model = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(class_weight="balanced"))
])

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "clf__C": [0.01, 0.1, 1, 10, 100]
}

grid = GridSearchCV(model, param_grid, scoring="f1", cv=5)
grid.fit(X_train, y_train)

print(grid.best_params_)
```

```python
y_pred = grid.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(model, X_sel, y, n_repeats=10, random_state=42)

perm_imp = pd.Series(result.importances_mean, index=X_sel.columns)
perm_imp.sort_values(ascending=False)
```
